```python
# Q8 

def decimal_to_twos_complement(decimal_number, num_bits):
    # Calculate the maximum positive value for the given number of bits
    max_positive_value = 2**(num_bits - 1) - 1

    # Check if the decimal number is within the representable range
    if decimal_number < -max_positive_value or decimal_number > max_positive_value:
        raise ValueError(f"Decimal number {decimal_number} out of representable range for {num_bits}-bit 2's complement.")

    # If the number is negative, convert it to its two's complement
    if decimal_number < 0:
        decimal_number += 2**num_bits

    # Convert the decimal number to binary
    binary_representation = bin(decimal_number)[2:].zfill(num_bits)

    return binary_representation

# Example usage:
decimal_number = -3
num_bits = 4
twos_complement_binary = decimal_to_twos_complement(decimal_number, num_bits)
print(twos_complement_binary)
```

2. Write a program to find if the given year is leap or not.

```python
def is_leap_year(year):
    # A year is a leap year if it is divisible by 4
    # But not divisible by 100 unless it is also divisible by 400
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Test the program with some examples
years = [2020, 1900, 2000, 2021]

for year in years:
    if is_leap_year(year):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")
```

3. A palindrom string is the one which reads same both forward and backward, i.e. ( "lal", "kaak", "01010").              
Write a program to find if given string is palindrome.


```python
def is_palindrome(input_string):
    # Convert the input string to lowercase to ignore case sensitivity
    input_string = input_string.lower()
    
    # Remove any spaces from the input string
    input_string = input_string.replace(" ", "")
    
    # Check if the input string is equal to its reverse
    if input_string == input_string[::-1]:
        return True
    else:
        return False

# Test the program with some examples
strings = ["lal", "kaak", "hello", "01010"]

for string in strings:
    if is_palindrome(string):
        print(f"{string} is a palindrome.")
    else:
        print(f"{string} is not a palindrome.")
```

Q4. Write a code to compute factorial of a given number.


```python
def factorial(n):
    # Base case: 0! and 1! are equal to 1
    if n == 0 or n == 1:
        return 1
    else:
        # Recursive case: factorial(n) = n * factorial(n-1)
        return n * factorial(n - 1)

# Test the program with some examples
num = [5, 0, 10]

for n in num:
    print(f"The factorial of {n} is: {factorial(n)}")
```

Q5. Write a code to compute greatest common divisor (GCD) of two integers a and b using Euclid's algorithm.

```python
def gcd(a, b):
    while b != 0:
        # Calculate the remainder when a is divided by b
        remainder = a % b

        # Swap a and b, so that a becomes the previous value of b, and b becomes the remainder
        a = b
        b = remainder

    # When b becomes 0, the GCD is found and is stored in 'a'
    return a

# Test the program with some examples
n1 = [36,100]
n2 = [48, 72]

for i,j in zip(n1,n2):
    print(f"The GCD of {i} and {j} is: {gcd(i,j)}")
```

Q6. Write a code to find if a number is prime or not.


```python
def is_prime(number):
    if number <= 1:
        return False
    elif number == 2:
        return True
    elif number % 2 == 0:
        return False
    else:
        # Check for factors from 3 to the square root of the number
        for i in range(3, int(number**0.5) + 1, 2):
            if number % i == 0:
                return False
        return True

# Test the program with some examples
num = [7, 12, 23, 49]

for n in num:
    if is_prime(n):
        print(f"{n} is a prime number.")
    else:
        print(f"{n} is not a prime number.")
```

1. Write a function to compute sum of digits of a number.

```python
def sum_of_digits(number):
    # Convert the number to a string to access individual digits
    num_str = str(number)
    
    # Initialize a variable to store the sum of digits
    sum_digits = 0
    
    # Iterate through each character in the string and add its integer value to the sum
    for digit in num_str:
        sum_digits += int(digit)
    
    return sum_digits

# Test the function with some examples
num = [12345, 9876, 0, 9999]

for n in num:
    print(f"The sum of digits of {n} is: {sum_of_digits(n)}")
```

2. Write a function to find the maximum number in an array.

```python
def find_max(arr):
    # Check if the array is empty
    if not arr:
        return None

    # Initialize a variable to store the maximum number
    max_num = arr[0]

    # Iterate through the array to find the maximum number
    for num in arr:
        if num > max_num:
            max_num = num

    return max_num

# Test the function with some examples
arrays = [[1, 5, 8, 3, 7, 2],
          [10, 20, 30, 40, 50],
          [-5, -2, -10, -3]]

for ar in arrays:
    print(f"The maximum number in {ar} is: {find_max(ar)}")
```

2. Make contour, density and surface plots for the electric potential of a line charge with charge density of $\lambda$ .  
Make appropriate scaling for the plot.                                                                                                                               
The electric potential (V) due to a line charge with charge density $$ at a distance 'r' from the line charge is given by:                                                                                                                           
$\qquad V=\frac{k}{r}$                                                                                                                                 
where:                                                                                                                                
    * $V$ is the electric potential.                                                                                                   
    * $k$ is Coulomb's constant $\displaystyle{\bigg(k=\frac{1}{4\pi\epsilon_o}}$ , where $\epsilon_o$ is the vacuum permittivity $\bigg).$                        
    * $$ is the charge density (charge per unit length) of the line charge.                                                              
    * $r$ is the distance from the line charge.

```python
# Charge density  (you can adjust this value)
lambda_ = 2.0

# Coulomb's constant
k = 8.9875e9  # N m^2/C^2

# Create a grid of points for the x and y coordinates
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the distance from the line charge at each point in the grid
r = np.sqrt(X**2 + Y**2)

# Calculate the electric potential V using the formula
V = k * lambda_ / r

# Plot the contour plot
plt.figure(figsize=(8, 6))
plt.contour(X, Y, V, levels=15, cmap='viridis')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Contour Plot of Electric Potential (V) of a Line Charge')
plt.colorbar(label='Electric Potential (V)')
plt.grid()
plt.show()

# Plot the density plot
plt.figure(figsize=(8, 6))
plt.imshow(V, extent=[-5, 5, -5, 5], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='Electric Potential (V)')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Density Plot of Electric Potential (V) of a Line Charge')
plt.show()

# Plot the surface plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, V, cmap='viridis', rstride=1, cstride=1)
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.set_zlabel('Electric Potential (V)')
ax.set_title('Surface Plot of Electric Potential (V) of a Line Charge')
plt.show()
```

The expectation values

The non-dimensionalized wave functions for these states are as follows:
* $1s$ state $\displaystyle{(n=1,l=0,m=0)\rightarrow _{1,0,0}(r,\theta,\phi)=\frac{1}{\sqrt{\pi}}e^{-r}}$                                              
* $2s$ state $\displaystyle{(n=2,l=0,m=0)\rightarrow _{2,0,0}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}(2-r)e^{-r/2}}$                                    
* $2p_x$ state $\displaystyle{(n=2,l=1,m=0)\rightarrow _{2,1,0}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}re^{-r/2}\cos(\theta)}$                                              
* $2p_y$ state $\displaystyle{(n=2,l=1,m=1)\rightarrow _{2,1,\pm 1}(r,\theta,\phi)=\frac{1}{\sqrt{32\pi}}re^{-r/2}\sin(\theta)e^{\pm i\phi}}$

To compute the expectation values, we need to integrate the product of the wave function and the corresponding operator over all space.

The radial expectation value $\langle r\rangle$ for a state $\psi_{nlm}$ is given by:                                                        
$$\displaystyle{\langle r\rangle \ =\int_0^\infty r^2 |\psi_{nlm}(r,\theta,\phi)|^2 dr}$$                                                                      
The radial squared expectation value $\langle r^2\rangle$ for a state $\psi_{nlm}$ is given by:                                              
$$\displaystyle{\langle r^2\rangle \ =\int_0^\infty r^4 |\psi_{nlm}(r,\theta,\phi)|^2 dr}$$                                                                   
The energy expectation value $\langle E\rangle$ for a state $\psi_{nlm}$ is given by:                                                        
$$\qquad\qquad\qquad\qquad\qquad\qquad\displaystyle{ \langle E\rangle \ =\int_0^\infty |\psi_{nlm}(r,\theta,\phi)|^2 \bigg(-\frac{1}{2} \frac{d^2}{dr^2} -\frac{1}{r}\frac{d}{dr}+\frac{l(l+1)}{2r^2}-\frac{Z}{r}\bigg)}$$                                                           
Where $Z$ is the atomic number (for hydrogen, $Z=1$).

```python
# Constants
Z = 1  # Atomic number for hydrogen (Z=1)

# Wave functions for the states
def psi_100(r, theta, phi):
    return 1 / np.sqrt(np.pi) * np.exp(-r)

def psi_200(r, theta, phi):
    return 1 / np.sqrt(32 * np.pi) * (2 - r) * np.exp(-r / 2)

def psi_210(r, theta, phi):
    return 1 / np.sqrt(32 * np.pi) * r * np.exp(-r / 2) * np.cos(theta)

def psi_211(r, theta, phi):
    return 1 / np.sqrt(64 * np.pi) * r * np.exp(-r / 2) * np.sin(theta) * np.exp(1j * phi)

# Radial expectation value <r>
def radial_expectation_value(psi):
    integrand = lambda r: r**2 * np.abs(psi(r, 0, 0))**2
    return si.quad(integrand, 0, np.inf)[0]

# Radial squared expectation value <r^2>
def radial_squared_expectation_value(psi):
    integrand = lambda r: r**4 * np.abs(psi(r, 0, 0))**2
    return si.quad(integrand, 0, np.inf)[0]

# Numerical second derivative of the wave function
def numerical_second_derivative(psi, r, h=1e-5):
    return (psi(r + h, 0, 0) - 2 * psi(r, 0, 0) + psi(r - h, 0, 0)) / h**2

# Energy expectation value <E>
def energy_expectation_value(psi):
    integrand = lambda r: np.abs(psi(r, 0, 0))**2 * (-1/2 * numerical_second_derivative(psi, r) - Z / r * psi(r, 0, 0))
    return si.quad(integrand, 1e-8, np.inf)[0]

# Compute the expectation values for each state
states = [(psi_100, "1s"), (psi_200, "2s"), (psi_210, "2p_x"), (psi_211, "2p_y")]

for psi, state_name in states:
    r_expectation = radial_expectation_value(psi)
    r_squared_expectation = radial_squared_expectation_value(psi)
    energy_expectation = energy_expectation_value(psi)

    print(f"State {state_name}:")
    print(f"<r>: {r_expectation}")
    print(f"<r^2>: {r_squared_expectation}")
    print(f"<E>: {energy_expectation}")
    print()
```

###  Lagrange interpolating polynomials in 2D

$\displaystyle L_j(x)=\sum_i\prod_{i',i'\ne j'} \frac{(x-x_{i'})}{(x_i-x_{i'})}\tilde{P}(x_i,y)  \qquad \tilde{P}(x_i,y)=\sum_j\prod_{j',j'\ne j} \frac{(y-y_{j'})}{(y_j-y_{j'})}f(x_i,y_j) \qquad P_n(x)=\sum_{i}\sum_{j} L_{i,j}(x,y)y_{i,j} $

```python
# Define the Lagrange basis polynomial for x
def lagrange_basis_x(x, i, x_values):
    product = 1
    for j in range(len(x_values)):
        if i != j:
            product *= (x - x_values[j]) / (x_values[i] - x_values[j])
    return product

# Define the Lagrange interpolating polynomial for x
def lagrange_interpolation_x(x, x_values, y_values, f_values):
    interpolated_value = 0
    for i in range(len(x_values)):
        basis_x = lagrange_basis_x(x, i, x_values)
        basis_y_sum = 0
        for j in range(len(y_values)):
            basis_y = lagrange_basis_y(y_values[j], j, y_values)
            basis_y_sum += basis_y * f_values[i][j]
        interpolated_value += basis_x * basis_y_sum
    return interpolated_value

# Define the Lagrange basis polynomial for y
def lagrange_basis_y(y, j, y_values):
    product = 1
    for j_prime in range(len(y_values)):
        if j != j_prime:
            product *= (y - y_values[j_prime]) / (y_values[j] - y_values[j_prime])
    return product
```

```python
def original_function(x, y):
    return np.exp(x + y)

# Choose points for interpolation along both x and y directions
num_points = [2, 3, 4]
x_values = np.linspace(0, 1, max(num_points))
y_values = np.linspace(0, 1, max(num_points))

# Create mesh grid for contour plots
X, Y = np.meshgrid(x_values, y_values)

# Initialize arrays for interpolated values
interpolated_values = np.zeros((len(x_values), len(y_values), len(num_points)))

# Calculate interpolated values using Lagrange interpolating polynomial
for n in range(len(num_points)):
    for i in range(len(x_values)):
        for j in range(len(y_values)):
            x = x_values[i]
            y = y_values[j]
            basis_x = np.array([lagrange_basis_x(x, k, x_values) for k in range(len(x_values))])
            basis_y = np.array([lagrange_basis_y(y, l, y_values) for l in range(len(y_values))])
            interpolated_values[i][j][n] = np.sum(basis_x * basis_y * original_function(x_values, y_values))

# Plot contour plots for interpolated polynomials along with the original function
plt.figure(figsize=(12, 4))

for n in range(len(num_points)):
    plt.subplot(1, 4, n+1)
    plt.contourf(X, Y, interpolated_values[:, :, n], levels=20, cmap='viridis')
    plt.colorbar()
    plt.title(f'Interpolated Polynomial (n={num_points[n]})')
plt.subplot(1, 4, 4)
plt.contourf(X, Y, original_function(X, Y), levels=20, cmap='viridis')
plt.colorbar()
plt.title('Original Function')
plt.tight_layout()
plt.show()
```

## Trapezoidal rule

${\displaystyle \int _{a}^{b}f(x)\,dx\approx (b-a)\cdot {\tfrac {1}{2}}\bigg(f(a)+f(b)\bigg).}$

```python
def trapz(f,a,b,N=50):
    x = np.linspace(a,b,N+1) 
    y = f(x)
    y_right = y[1:] 
    y_left = y[:-1] 
    dx = (b - a)/N
    T = (dx/2) * np.sum(y_right + y_left)
    return T
```

## Simpson's 1/3 rule

${\displaystyle \int _{a}^{b}f(x)\,dx\approx {\frac {b-a}{6}}\left[f(a)+4f\left({\frac {a+b}{2}}\right)+f(b)\right].}$

```python
def simps_1_3_rule(f,a,b,N=50):
    if N % 2 == 1:
        raise ValueError("N must be an even integer.")
    dx = (b-a)/N
    x = np.linspace(a,b,N+1)
    y = f(x)
    S = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S
```

## Simpson's 3/8 rule                                                                              
$\qquad\displaystyle \int _{a}^{b}f(x)\,dx \approx {\frac {3}{8}}h\left[f(a)+3f\left({\frac {2a+b}{3}}\right)+3f\left({\frac {a+2b}{3}}\right)+f(b)\right]$                                                                               
$\qquad\qquad\qquad\quad\displaystyle ={\frac {b-a}{8}}\left[f(a)+3f\left({\frac{2a+b}{3}}\right)+3f\left({\frac {a+2b}{3}}\right)+f(b)\right] $                                                                              
where ${\displaystyle h=\frac{(b-a)}{3}}$ is the step size.                                                                              
The error of this method is                                                                              
$\qquad{\displaystyle -{\frac {3}{80}}h^{5}f^{(4)}(\xi )=-{\frac {(b-a)^{5}}{6480}}f^{(4)}(\xi ),}$                                                                              
where $\xi$  is some number between $a$ and $b$.

```python
def simpsons_3_8_rule(func, a, b, n):
    if n % 3 != 0:
        raise ValueError("Number of intervals must be a multiple of 3.")
    
    h = (b - a) / n
    integral = func(a) + func(b)
    
    for i in range(1, n):
        x = a + i * h
        if i % 3 == 0:
            integral += 2 * func(x)
        else:
            integral += 3 * func(x)
    
    integral *= (3 * h) / 8
    return integral
```

### Gauss Quadrature

```python
from scipy.special import roots_legendre as rl

def gauss(f,a,b,N=50):
    V = rl(N)
    t = V[0]
    w = V[1]
    x = ((b-a)/2)*t + (b+a)/2
    k = ((b-a)/2)*f(x)
    G = np.sum(w*k)
    display(pd.DataFrame({'weights $W_i$':w,
                          'Roots $X_i$':x,
                          '$f(x_i)$': f(x),
                          '$w_if(x_i)$': w*k}))
    return G
```

$\displaystyle{f(x) \approx P_2(x)=\frac{x-x_{i+1}}{x_i-x{i+1}}f_i + \frac{x-x_i}{x_{i+1}-x_i}f_{i+1}}$

$\displaystyle\bf\text{Forward Difference } :{f'(x_i) \approx P'_2(x_i)=D_{+}f=\frac{f_{i+1}-f_{i}}{h_i}}$

$\displaystyle\bf\text{Backward Difference } :{f'(x_{i+1}) \approx P'_2(x_{i+1})=D_{-}f=\frac{f_{i+1}-f_{i}}{h_i}}$

$\qquad\qquad\qquad\qquad\qquad\qquad\qquad\displaystyle{3-Points } :$

$\displaystyle{f(x) \approx P_3(x)=\frac{(x-x_{i})(x-x_{i+1})}{(h_{i-1}+h_i)(h_{i-1})}f_{i-1} + \frac{(x-x_{i-1})(x-x_{i+1})}{(h_{i-1})(-h_i)}f_{i}+\frac{(x-x_{i-1})(x-x_{i})}{(h_{i-1}+h_i)(h_i)}f_{i+1}}$

$\displaystyle\bf\text{Forward Difference } :{f'(x_{i-1}) \approx P'_3(x_{i-1})=D_{+}f=\frac{-3f_{i+1}+4f_{i}-f_{i+1}}{2h}}$

$\displaystyle\bf\text{Backward Difference } :{f'(x_{i+1}) \approx P'_3(x_{i+1})=D_{-}f=\frac{f_{i-1}-4f_{i}+3f_{i+1}}{2h}}$

$\displaystyle\bf\text{Central Difference } :{f'(x_{i}) \approx P'_3(x_{i})=\frac{1}{2}(D_{-}+D_{+})f=\frac{f_{i+1}-f_{i-1}}{2h}}$

```python
def FD2(f, x, h):
    return (f(x + h) - f(x)) / h

def BD2(f, x, h):
    return (f(x) - f(x - h)) / h

def CD3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def FD3(x, h):
    return (-3*f(x) + 4*f(x+h) - f(x+2*h)) / (2*h)

def BD3(x, h):
    if x - 2 * h == 0:
        print("Function is undefined at x =", x - 2 * h)
        return None
    return (f(x-2*h) - 4*f(x-h) + 3*f(x)) / (2*h)
```

Using a large number of points (8-10 or more) for derivative computations, especially in finite difference methods, can lead to several issues:

Higher Sensitivity to Noise: As you increase the number of points, the finite difference methods become more sensitive to noise in the data.                                                                                                                                     
Small variations in the function values can result in large fluctuations in the computed derivatives,                                        
leading to less accurate results.

Numerical Instability: Finite difference methods involve subtracting or adding small values to the function evaluations.                  
When subtracting two nearly equal values, you can experience a significant loss of precision due to subtractive cancellation,                                                                                                                                      
leading to numerically unstable results.

Round-off Errors: When subtracting or adding small values to a large value, the relative round-off error can become significant.                                                                                                                                  
This can introduce additional errors in the derivative computations.

Higher Computational Cost: Calculating derivatives using a larger number of points requires more function evaluations and computations,                                                                                                                                       
increasing the computational cost and execution time.

Inaccurate Results Near Discontinuities: Finite difference methods assume a certain degree of smoothness in the function.                  
If you're working with functions that have discontinuities or sharp changes, using more points can lead to inaccurate derivatives.

Higher Order Methods: When you need accurate derivatives, it's often more effective to use higher-order methods                          
(e.g., 5-point or 7-point stencils) instead of increasing the number of points in a lower-order method.                                                  
Higher-order methods offer better accuracy without the issues associated with excessive points.

Limited Interval Size: As you increase the number of points, the interval size h (the step size) becomes smaller.                                                           
This can lead to numerical issues because the floating-point precision may not be sufficient to accurately represent the small differences in function values.

To address these issues, it's generally recommended to use appropriate higher-order finite difference formulas or consider alternative numerical differentiation techniques like Richardson extrapolation, spline interpolation, or using methods like the Savitzky-Golay filter for derivative estimation. ### Euler's Forward Method  or, explicit method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n)} + (t) f(x^{(n)},t_{n})   .\end{equation*}$

```python
def euler_forward(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1     # (n-1) divisions
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(n-1):
        x[i+1] = x[i] + f(x[i], t[i])*dt
    return t,x
```

### Simpson's 1/3 Method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n-1)} + (1/3) (t)[ f(x^{(n-1)},t_{n-1}) + 4f(x^{(n)},t_{n}) +  f(x^{(n+1)},t_{n+1})]  .\end{equation*}$

### Simpson's 3/8 Method

```python
def simpsons_3_8(f, t_init, t_fin, dt, init_condition):    
    n = int((t_fin - t_init) / dt) + 1
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(n - 3):
        k1 = f(x[i], t[i])
        k2 = f(x[i] + (dt / 3) * k1, t[i] + dt / 3)
        k3 = f(x[i] + (2 * dt / 3) * k2, t[i] + 2 * dt / 3)
        k4 = f(x[i] + dt * k1 - 2 * dt * k2 + 3 * dt * k3, t[i] + dt)
        x[i+1] = x[i] + (dt / 8) * (k1 + 3 * k2 + 3 * k3 + k4)
    return [t[:-2], x[:-2]]
```

2. Why do the numerical solutions of ODEs have errors ?                                                                                                                               
$\text{Ans }:$ Numerical solutions of ordinary differential equations (ODEs) have errors due to several factors, both inherent to the nature of numerical methods and due to the limitations of computers.                                                                                                                                
Here are some of the main reasons for the errors in numerical solutions of ODEs:                                                                                                                               
$\bf\text{Discretization Error:}$ Numerical methods convert continuous ODEs into discrete equations by dividing the continuous domain into discrete time steps or grid points.                                                                                                                                
This discretization introduces errors because it approximates the continuous behavior of the system.                                                                                                                               
$\bf\text{Truncation Error:}$ Numerical methods use finite approximations of derivatives, leading to truncation errors. The accuracy of these approximations depends on the chosen step size and the order of the method.                                                                                                                               
$\bf\text{Round-off Error:}$ Computers use finite-precision arithmetic, which means that real numbers are approximated using a finite number of bits. This can lead to round-off errors where calculations involve approximated values, especially in cases of repeated calculations or very large/small numbers.                                                                                                                               
$\bf\text{Stability Issues:}$ Some numerical methods, particularly explicit methods like the Euler method, have stability constraints. If the step size is too large, these methods can lead to numerical instability, causing solutions to diverge from the true solution.                                                                                                                               
$\bf\text{Initial Condition Error:}$ The initial condition used in numerical methods might not be exactly the same as the actual initial condition of the ODE. Even small deviations can lead to errors in the subsequent calculations.                                                                                                                               
$\bf\text{Accumulated Errors:}$ Numerical errors can accumulate over time as the calculation progresses through multiple steps. Small errors in each step can compound, leading to larger discrepancies from the true solution.                                                                                                                               
$\bf\text{Sensitivity to Parameters:}$ Some ODEs are highly sensitive to changes in parameters or initial conditions. Small variations in these values can lead to significantly different behaviors in the solution.                                                                                                                               
$\bf\text{Numerical Method Selection:}$ Different numerical methods have different accuracy and stability properties. Choosing an inappropriate method for a specific problem can lead to larger errors.                                                                                                                               
$\bf\text{Stiffness:}$ Some ODEs exhibit stiffness, where the system's behavior changes rapidly in some regions and slowly in others. Numerical methods can struggle to accurately capture both behaviors.                                                                                                                               
$\bf\text{Nonlinearity:}$ Nonlinear ODEs can be challenging to solve numerically because small changes in the initial conditions or parameters can lead to complex and unexpected behaviors.                                                                                                                               
Due to these factors, it's important to choose appropriate numerical methods, adjust step sizes, and consider the specific characteristics of the problem to achieve accurate numerical solutions. Additionally, error analysis and comparisons with analytical solutions are valuable for understanding the accuracy and reliability of numerical results.

```python
def euler_backward(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1
    t_values=np.linspace(t_init, t_fin, n)
    x_values = np.zeros(n)
    x_values[0] = init_condition
    for i in range(1, n):
        x_next = x_values[i - 1] + dt * f(x_values[i - 1], t_values[i - 1])
        for _ in range(10):  # Use Newton-Raphson method
            x_new = x_values[i - 1] + dt * f(x_next, t)
            if abs(x_new - x_next) < 1e-6:
                break
            x_next = x_new
        x_values[i] = x_next 
    return t_values, x_values
```

### Trapezoidal Method      or ,      semi-implicit method
$\qquad\begin{equation*} x^{(n+1)} = x^{(n)} + (0.5)(t) [f(x^{(n)},t_{n}) + f(x^{(n+1)},t_{n+1})]  .\end{equation*}$

```python
# Trapezoid method implementation
def trapezoid_method(f, t_init, t_fin, dt, init_condition):
    n = int((t_fin - t_init) / dt) + 1
    t = np.linspace(t_init, t_fin, n)
    x = np.zeros(n)
    x[0] = init_condition
    for i in range(1, n):
        t_i = t[i - 1]
        x_i = x[i - 1]
        x_new = x_i + 0.5 * dt * (f(x_i, t_i) + f( x_i + dt * f(x_i,t_i),t_i + dt))   # update x
        x[i] = x_new
    return t, x
```

###  Predictor-Corrector Scheme
$ \text{Predictor step}: x^{(n+1)^*} = x^{(n)}+(\Delta t) f( x^{(n)}, t_n)  $                                                                            
$\text{Corrector step}:  x^{(n+1)} = x^{(n)}+() (\Delta t) [f( x^{(n)}, t_n)+f( x^{(n+1)*}, t_{n+1})]$

```python
def PC(f, tinit, tfinal, dt, init_cond):
    n=int((tfinal-tinit)/dt)+1
    t = np.linspace(tinit,tfinal,n)
    x = np.zeros(n)
    x[0] = init_cond
    
    for i in range(1, n):
        x_pred = x[i - 1] + dt * f(x[i - 1], t[i - 1])
        x[i] = x[i - 1] + 0.5 * dt * (f(x[i - 1], t[i - 1]) + f(x_pred, t[i]))
    return t, x
```

### Second-order Runge-Kutta scheme (RK2)

$\text{The predictors for RK2 is} $                                                                            
$\qquad k_1 = (\Delta t) f(x^{(n)},t_n); \quad x^{(n+)^*} = x^{(n)} + k_1/2 ,$                                                                            
$\text{Now, the corrector for RK2 is } $                                                                            
$\qquad x^{(n+1)} = x^{(n)} + (t) f(x^{(n+)^*},t_{n+}) $

```python
def RK2(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = init_cond

    for i in range(n - 1):
        k1 = dt * f(x[i], t[i])  
        x[i + 1] = x[i] + dt * f(x[i] + k1/2, t[i] + dt/2)
    return t, x
```

### Fourth-order Runge-Kutta  scheme (RK4)

$\text{The predictors for RK4 are} $                                                                            
$\qquad k_1 = (\Delta t) f(x^{(n)},t_n);  \quad\qquad x^{(n+)^*} = x^{(n)} + k_1/2 , $                                                                            
$\qquad k_2 = (\Delta t) f(x^{(n+)^*},t_{n+}); \  \ x^{(n+)^{**}} = x^{(n)} + k_2/2 ,$                                                                            
$\qquad k_3 = (\Delta t) f(x^{(n+)^{**}},t_{n+}); \ \ x^{(n+1)^{***}} = x^{(n)} + k_3 , $                                                                            
$\qquad k_4 = (\Delta t) f(x^{(n+1)^{***}},t_{n+1}) .$                                                                            
$\text{Now, the corrector for RK4 is } $                                                                                                            
$\qquad x^{(n+1)} = x^{(n)} + (k_1 + 2k_2 +2k_3 + k_4)/6 $

```python
def RK4(f, tinit, tfinal, dt, init_cond):
    n=int((tfinal-tinit)/dt)+1
    t = np.linspace(tinit,tfinal,n)
    x = np.zeros(n)
    x[0] = init_cond
    for i in range( n-1 ):
        k1=f(x[i],t[i])
        k2=f(x[i]+(dt*k1/2),t[i]+(dt/2))
        k3=f(x[i]+(dt*k2/2),t[i]+(dt/2))
        k4=f(x[i]+dt*k3,t[i]+dt)
        x[i+1] = x[i] + (dt/6)*(k1 + 2*(k2 + k3) + k4)
    return [t, x]
```

### Leap-frog method
$\begin{equation*}\displaystyle{x^{(n+1)} = x^{(n1)}+2(\Delta t) f( x^{(n)}, t_n)}\end{equation*}$

```python
def LF(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = init_cond
    
    for i in range(1, n):
        x_half = x[i - 1] + 0.5 * dt * f(x[i - 1], t[i - 1])  
        x[i] = x[i - 1] + dt * f(x_half, t[i - 1])  
    return t, x
```

### Second-order Adams-Bashforth method
$\displaystyle{x^{(n+1)} = x^{(n)} + (\Delta t) f( x^{(n)}, t_n) ,}$                                                                            

$ \displaystyle{x^{(n+1)} = x^{(n)} + \frac{(\Delta t)}{2} [3 f( x^{(n)}, t_n)  f( x^{(n1)}, t_{n1})]}$

```python
def AB2(f, tinit, tfinal, dt, init_cond):
    n = int((tfinal - tinit) / dt) + 1    
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)    
    x[0] = init_cond
    
    for i in range(1, n):
        x_e = x[i-1] + dt * f(x[i-1], t[i])  
        if i > 1:
            x_ab2 = x[i-1] + (dt / 2) * (3 * f(x[i-1], t[i]) - f(x[i-2], t[i-2]))
        else:
            x_ab2 = x_e        
        x[i] = x_ab2
    return t, x
```

$\displaystyle{x^{(n+1)} = x^{(n)} + \frac{(\Delta t)}{12} [5f( x^{(n+1)}, t_{n+1}) + 8f( x^{(n)}, t_{n}) - f( x^{(n-1)}, t_{n-1})]}$

```python
def AM3(f, tinit, tfinal, dt, initcond):
    n = int((tfinal - tinit) / dt) + 1
    t = np.linspace(tinit, tfinal, n)
    x = np.zeros(n)
    x[0] = initcond
    for i in range(2):  # Use a higher-order method to start (e.g., RK4)
        k1 = dt * f(x[i], t[i])
        k2 = dt * f(x[i] + k1 / 2, t[i] + dt / 2)
        k3 = dt * f(x[i] + k2 / 2, t[i] + dt / 2)
        k4 = dt * f(x[i] + k3, t[i] + dt)
        x[i + 1] = x[i] + (k1 + 2 * k2 + 2 * k3 + k4) / 6
    
    for i in range(2, n - 1):
        x_pred = x[i] + (dt / 12) * (5 * f(x[i], t[i]) + 8 * f(x[i - 1], t[i - 1]) - f(x[i - 2], t[i - 2]))
        x[i + 1] = x[i] + (dt / 12) * (5 * f(x_pred, t[i + 1]) + 8 * f(x[i], t[i]) - f(x[i - 1], t[i - 1]))
    return t, x
```

d) $\dot{x}=x^250x$  with x(0) = 10.                                                                                                               
$\displaystyle \text{Separate variables }:$
                                                                                                               
$\displaystyle\quad \frac{dx}{dt} = x^2 - 50x $
                                                                                                               
$\displaystyle or,\quad \frac{dx}{x^2 - 50x} = dt $
                                                                                                               
$\displaystyle \text{Now, we can integrate both sides }:$
                                                                                                               
$\displaystyle or,\quad\int \frac{1}{x^2 - 50x} dx = \int dt $
                                                                                                               
$\displaystyle \text{To integrate the left-hand side, we can use partial fraction decomposition }:$
                                                                                                               
$\displaystyle or,\quad\int \frac{1}{x(x - 50)} dx = \int dt $
                                                                                                               
$\displaystyle\text{Now, we can write it as two separate integrals }:$
                                                                                                               
$\displaystyle or,\quad\int \left(\frac{A}{x} + \frac{B}{x - 50}\right) dx = \int dt $
                                                                                                               
$\displaystyle\text{Solving for }A \text{ and }B \text{ common denominator is }x(x - 50) : $
                                                                                                               
$\displaystyle\quad A(x - 50) + Bx = 1 $
                                                                                                               
$\displaystyle\text{Now, let's find the values of }A \text{ and }B :$
                                                                                                               
$\displaystyle\text{Setting }\big(x = 0\big) , \text{we get }:$

$\displaystyle\text{Now, we can integrate the equation }: $
                                                                                                               
$\displaystyle or,\qquad -\frac{1}{50} \int \frac{1}{x} dx + \frac{1}{50} \int \frac{1}{x - 50} dx = \int dt $

$\displaystyle or,\qquad \ln\left|\frac{x - 50}{x}\right| = 50t + C $
                                                                                                               
$\displaystyle\text{Now, we can remove the absolute value by taking the exponential of both sides }:$
                                                                                                               
$\displaystyle or,\qquad \frac{x - 50}{x} = \exp({50t + C})$
                                                                                                               
$\displaystyle\text{Now, we can find the value of }C\text{ using the initial condition }\big(x(0) = 10 \big): $

```python
def leap_frog2D(f, tinit, tfinal, dt, initcond):
    n = int((tfinal-tinit)/dt)+1 # n divisions
    t = np.linspace(tinit,tfinal,n)
    y = np.zeros((n,2))
    y[0][0] = initcond[0]; # position
    y[0][1] = initcond[1] + f(y[0][0],y[0][1])*dt/2
    # init vel at t=1/2
    for k in range(1,n):
        y[k][0] = y[k-1][0] + y[k-1][1]*dt
        y[k][1] = y[k-1][1] + f(y[k][0],y[k][1])*dt
    return t, y
```

To find the amplitude of the Fourier mode $\big(\mathbf{k} = 1\big)$ for the function $\big(f(x) = 4\cos(x) + 2\sin(2x)\big)$,                                                                                                                                
we need to calculate the coefficients $a_1$ and $b_1$ in the Fourier series representation.                                                            
The amplitude of the Fourier mode $\big(\mathbf{k}\big)$ is given by:                                                                                                                               
$\displaystyle A_{\mathbf{k}} = \sqrt{a_{\mathbf{k}}^2 + b_{\mathbf{k}}^2}$                                                                                                                               
For $\big(\mathbf{k} = 1\big)$ , the coefficients $a_1$ and $b_1$ are calculated as follows:                                                                                                                               
$\quad\displaystyle a_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos(x) + 2\sin(2x)\right) \cos(x) dx$                        
$\quad\displaystyle b_1 = \frac{1}{\pi} \int_{0}^{2\pi} \left(4\cos(x) + 2\sin(2x)\right) \sin(x) dx$                                                                                                                               
Let's calculate these integrals:

### Spectral Method

$\textbf{Diffusion equation}$ is

$\qquad \displaystyle \frac{\partial \phi}{\partial t} = \kappa \frac{\partial^2\phi}{\partial t^2}$

The diffusion equation in Fourier space is

$\qquad \displaystyle \frac{\partial \phi_k}{\partial t} = -\kappa k^2 \hat{\phi}_k  $

which solution is

$\qquad \displaystyle \hat{\phi}_k(t)=\hat{\phi}_k(t=0)\cdot\exp(-\kappa k^2 t)$

For the initial condition, $\phi(x, t=0),$ we can compute

$\qquad \displaystyle \hat{\phi}_k\big(t=0\big) = \frac{1}{L} \int^L_0 \phi\big(x,t=0\big)\cdot \exp(-ikx) \ dx$

The Fourier transform of the initial condition $\phi\big(t=0\big) = \exp\big(2(x\pi)^2\big)$ is

$\qquad \displaystyle \hat{\phi}\big(k,t=0\big) = a\cdot\exp[-\frac{1}{2}(ka)^2]\cdot\exp(-ika)$  where $a=\frac{1}{2}$  

Hence,

$\qquad \displaystyle \hat{\phi}\big(k,t\big) =a\cdot\exp[-\frac{1}{2}(ka)^2-\kappa k^2 t]\cdot\exp(-ika)$

whose inverse transform yields

$\qquad \displaystyle \hat{\phi}\big(k,t\big) = \frac{a}{a'}\exp\big[-\frac{1}{2{a'}^2}(x-\pi)^2\big]$ where $\displaystyle{a' = \sqrt{(a^2+\kappa t)}}$  at  $t=2$.

```python
kappa = 1
tf = 1
L = 2*np.pi
N = 64
h = L/N
j = np.arange(0,N)
x = j*h

f = np.exp(-2*(x-np.pi)**2)  # init cond
plt.plot(x,f,label='t=0')

a = 1/2

def inverse_fourier_transform(kappa, t):
    a_prime = np.sqrt(a**2 + kappa * t)
    return (a / a_prime) * np.exp(-(0.5 / a_prime**2) * (x - np.pi) ** 2)

plt.plot(x, inverse_fourier_transform(kappa, t=2),'r-',label='Exact')

fk = np.fft.rfft(f,N)/N
kx = np.linspace(0, N//2, N//2+1)
fk_t = fk*np.exp(-kappa*kx**2*tf)  
f_t = np.fft.irfft(fk_t,N)*N  # in Real space
plt.plot(x,f_t,'k--',label='t=1')

plt.title('Diffusion equation : ' + \
          r'$\displaystyle{\frac{\partial \phi}{\partial t} = \kappa \frac{\partial^2\phi}{\partial t^2}} $',
          fontsize=20,y=1.1)
q1=[0,np.pi,(2*np.pi)]
q2=[r'$0$',r'${\pi}$',r'${2\pi}$']
plt.xticks(q1,q2)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
plt.show()
```

$or,\qquad\displaystyle \frac{\partial u}{\partial t}=-c\frac{\partial u}{\partial x}$

initial condition $\qquad\displaystyle \hat{\phi}_k(t)=\hat{\phi}_k(t=0)\exp(-ikct)$

Euler Forward method implimentation                                                                                                                                      
$\qquad\displaystyle \hat{\phi}_k^{n+1} = \big[1-ick\Delta t\big] \hat{\phi}_k^{n} $

```python
c = 1.0   # Wave speed
L = 10.0  # Length of the domain
N = 512   # Number of grid points
T = 5.0   # Total simulation time
dt = 0.01 # Time step
x = np.linspace(0, L, N, endpoint=False)
dx = x[1] - x[0]
phi_0 =  np.exp(-2*(x-np.pi)**2) 

phi = phi_0.copy()
phi_new = np.zeros(N)
for t in np.arange(0, T , dt):
    if t == 0:
        plt.plot(x, phi, label=f't = {t:.2f}')
    k = 2 * np.pi * np.fft.fftfreq(N, d=dx)  # Wave numbers
    phi_hat = np.fft.fft(phi)
    phi_new_hat = phi_hat * np.exp(-1j * k * c * dt)
    phi_new = np.fft.ifft(phi_new_hat).real
    phi, phi_new = phi_new, phi
    
plt.plot(x, phi, label=f't = {t:.2f}')
plt.xlabel('x')
plt.ylabel('phi(x, t)')
plt.title(r'Wave Equation : $\displaystyle \frac{\partial u}{\partial t}=-c\frac{\partial u}{\partial x}$',
           fontsize=20,y=1.1)
plt.legend()
plt.grid()
plt.show()
```

### Burgers equation

Burgers equation is a flow equation for a fully-compressible fluid:

$\qquad\displaystyle \frac{\partial u}{\partial t}+u\frac{\partial u}{\partial x}=\nu\frac{\partial^2 u}{\partial x^2}$

$or,\qquad\displaystyle \frac{\partial u}{\partial t}+\frac{1}{2}\frac{\partial u^2}{\partial x}=\nu\frac{\partial^2 u}{\partial x^2}$                                                                                                                                                                                                     
$or,\qquad\displaystyle \frac{\partial u}{\partial t}=-\frac{1}{2}\frac{\partial u^2}{\partial x}+\nu\frac{\partial^2 u}{\partial x^2}$

n Fourier space, the equation for the Fourier mode with wavenumber $k$ is

$\qquad\displaystyle \frac{d}{dt}\hat{u}_k=-i\frac{k}{2}\hat{u}_k^2-\nu k^2 \hat{u}_k=-\widehat{N}_k-\nu k^2 \hat{u}_k$

```python
nu = 0.1
tf = 1.0
dt = 0.02
N = 512
L = 2 * np.pi
h = L / N
j = np.arange(0, N)
x = j * h

kx = np.linspace(0, N // 2, N // 2 + 1)
exp_factor_dtby2 = np.exp(-nu * kx**2 * dt / 2)
exp_factor = np.exp(-nu * kx**2 * dt)

def comput_Nk(fk):
    f = np.fft.irfft(fk, N) * N
    f = f * f
    fk_prod = np.fft.rfft(f, N) / N
    return (1j * kx * fk_prod)

f = np.sin(x)  # Initial condition
fk = np.fft.rfft(f, N) / N  # FT(f)

time_steps = []
profiles = []

nsteps = int(tf / dt)
for i in range(nsteps + 2):
    if i % 10 == 0:  # Store every 10 time steps for plotting
        f_profile = np.fft.irfft(fk, N) * N
        time_steps.append(i * dt )
        profiles.append(f_profile)

    Nk = comput_Nk(fk)
    fk_mid = (fk - (dt / 2) * Nk) * exp_factor_dtby2
    Nk_mid = comput_Nk(fk_mid)
    fk = (fk - dt * Nk_mid) * exp_factor
    fk_t = fk * exp_factor_dtby2
    
q1=[0,np.pi,(2*np.pi)]
q2=[r'$0$',r'${\pi}$',r'${2\pi}$']
plt.figure(figsize=(8, 4))
for i, t in enumerate(time_steps):
    plt.plot(x, profiles[i], label=f'Time = {t:.2f}')
plt.xticks(q1,q2)
plt.xlabel('x')
plt.ylabel(r'$\phi$')
plt.title('Burgers Equation :'+ \
    r' $\displaystyle\frac{\partial u}{\partial t}=-u\frac{\partial u}{\partial x}+\nu\frac{\partial^2 u}{\partial x^2}$',
          fontsize=20,y=1.1)
plt.grid()
plt.legend()
plt.show()
```

### Navier-Stokes Equation

$\qquad \quad \displaystyle \frac{\partial \mathbf{u}}{\partial t}+\mathbf{u}\cdot\nabla \mathbf{u}=-\nabla p+\nu\nabla^2 \mathbf{u}$                                                                                                                                                                                                           
$\qquad \quad \displaystyle\nabla \cdot\mathbf{u} = 0$

$\qquad \quad$where $\mathbf{u} , p $ are the velocity and pressure fields respectively, and $\nu$ is the kinematic viscosity.

$\qquad \quad \displaystyle \frac{\partial }{\partial t} \hat{u}_{\mathbf{k},i} = -ik_j \widehat{u_j(\mathbf{r})u_i(\mathbf{r})}_\mathbf{k} - ik_i\hat{p}_\mathbf{k}-\nu k^2\hat{u}_{\mathbf{k},i}$

$\qquad \quad \displaystyle \frac{\partial }{\partial t} \hat{u}_{\mathbf{k},i} = -N_{\mathbf{k},i} - ik_i\hat{p}_\mathbf{k}-\nu k^2\hat{u}_{\mathbf{k},i}$

$\qquad \quad \displaystyle k_i\hat{u}_{\mathbf{k},i}=0$

Here, $\hat{p}_k$ is the Fourier transform of the pressure field, and $\hat{N}_{k, i}$ is Fourier transform of the non-linear term $(u_j\partial_ju_i)$.                                                                                                                                                               
Note that $i$ in front of $\hat{N}_{k,i}$ is $\sqrt{1}$.                                                                                                                         
We divide the real space so that there are $M = N_1\times N_2\times N_3$ grid points.                                                                                                                                          
The wavenumber components of the modes are $(2/L_i)n_i$,                                                                                                                                          
where $n_i$ takes values from $[N_1/2+1:N_1/2,N_2/2+1:N_2/2, 0:N_3/2+1]$. 

The nonlinear term is computed using the following steps

1. compute u(x) using ifft($\hat{u}_\mathbf{k}$)

2. compute products $u_iu_j$

3. compute the Fourier transform of $u_iu_j$ for all combinations

4. multiply $\text{FT}(u_iu_j)$ by $ik_j$ that yields $ ik_j \widehat{u_j(\mathbf{r})u_i(\mathbf{r})}_\mathbf{k'}$ which is the Fourier transform of the non-linear term.                                                                                                                                   
This computation requires $O(M\log(M))$ operations.


$\qquad \quad \displaystyle -\nabla p = \nabla\cdot[u\cdot\nabla u]=\nabla\cdot N$   whose Fourier transform is $\quad  \displaystyle{\hat{p}_\mathbf{k}=i\frac{1}{k^2} \mathbf{k}\cdot\mathbf{\hat{N}_k} }$

Eulers forward method

$\qquad \quad \displaystyle \mathbf{\hat{u}_k}^{(n+1)} = \mathbf{\hat{u}_k}^{(n)}[1-\nu k^2 (\Delta t)]+(\Delta t)[-\hat{N}_\mathbf{k}-i\mathbf{k}\hat{p}_\mathbf{k}^{(n)}]$

```python
L = 1.0      # Domain size
N = 24       # Number of grid points in each dimension
nu = 1e-4    # Kinematic viscosity
u_rms = 1.0  # Root-mean-square velocity
Re = u_rms * L / nu
tau_nu = Re / N
tau_NL = Re**(-1/4) * tau_nu
dt = 0.01    # Time step
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
X, Y = np.meshgrid(x, y)
u = np.sin(2 * np.pi * X / L) * np.cos(2 * np.pi * Y / L)
v = -np.cos(2 * np.pi * X / L) * np.sin(2 * np.pi * Y / L)

plt.figure(figsize=(8, 6))
plt.quiver(X, Y, u, v, scale=10, color='b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Velocity Field')
plt.show()
```
